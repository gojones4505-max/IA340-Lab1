### The section I highlighted was the area in which each case occured. I then spliced the data to view only the 5-10 rows of data. I also worted the data by the crime description and also showed the data in which the crime happened at a time greated than 10:00.

```python
sum_Vict_Age_by_Area = df.groupby('Vict Age')['AREA'].sum() # calculate the total deaths per day
sum_Vict_Age_by_Area.nlargest(10).plot.bar() # use a bar chart to show the results of the top 10 countries 
```

### This graph shows the Victims age by the area number where the crime occured. This shows the highest age from incidents in each area.

```python
sum_Premis_Desc_by_time_occ = df.groupby('Premis Desc')['TIME OCC'].sum() #aggregate the total case number per day
```
