### Importing Secret Manager Function

```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
    
```

### Question 1

```python
%%ai gpt4 -f code
Calculate the growth rate of the yearly population with the data provided in {me_df}, use the me_df directly, and visualize the result on a line chart that includes the state name in the title
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Given DataFrame
me_df = pd.DataFrame({
    'fips': [23]*15,
    'income': [42801, 43439, 45888, 46581, 45734, 45815, 46033, 46709, 46974, 49462, 51494, 53079, 56277, 55602, 58924],
    'population': [1283673, 1321574, 1317207, 1316456, 1318301, 1327567, 1328188, 1329192, 1328302, 1330089, 1329328, 1331479, 1335907, 1338404, 1344212],
    'year': range(2005, 2020)
})

# Calculate the Year-on-Year Growth Rate
me_df['pop_growth_rate'] = me_df['population'].pct_change() * 100

# Plotting the chart
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(me_df['year'], me_df['pop_growth_rate'], color='tab:blue')
ax.set_xlabel('Year')
ax.set_ylabel('Population Growth Rate (%)')
ax.set_title('Yearly Population Growth Rate of Maine')

plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sort the dataframe by year
me_df = me_df.sort_values('year')

# Set up the figure and axis
fig, ax1 = plt.subplots(figsize=(10,6))

color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Population', color=color)
ax1.plot(me_df['year'], me_df['population'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Income', color=color)  
ax2.plot(me_df['year'], me_df['income'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title('Population and Median Household Income of Maine over the years')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Calculate income growth rate
me_df['income_growth_rate'] = me_df['income'].pct_change() * 100

# Compute average income rate per year
average_income_rate_per_year = me_df.groupby('year')['income_growth_rate'].mean()

# Plot the bar chart
plt.figure(figsize=(10,6))
plt.bar(average_income_rate_per_year.index, average_income_rate_per_year.values, color='green')

plt.xlabel('Year')
plt.ylabel('Average Income Growth Rate (%)')
plt.title('Average Income Growth Rate Per Year for Maine')

plt.show()
```
